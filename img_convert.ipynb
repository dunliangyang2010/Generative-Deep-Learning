```python
from PIL import Image
content_img_path = '/content/000293.jpg'

def read_img(img_name):
  img = Image.open(img_name)
  np_img = np.array(img)
  np_img = np_img.astype('float32') / 255.
  
  return np_img

content_val = read_img(content_img_path)
print(content_val.shape) # 注意读取出的图片长宽顺序调换了
```

```python
print(type(content_val))
```

```python
ima = Image.fromarray(content_val)
ima.save("filename1.jpeg")
```
